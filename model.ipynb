```python
def meanRnage(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1]))/2
    try:
        return float(x)
    except :
        return None
```

```python
def remove_sqft_outlier(df):
    df_output = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_feet)

        st = np.std(subdf.price_per_feet)
        gen_df = subdf[(subdf.price_per_feet > (m-st)) & (subdf.price_per_feet <= (m+st))]
        df_output = pd.concat([df_output,gen_df],ignore_index=True)
    return df_output
data = remove_sqft_outlier(data)
data.describe()
```

```python
def bhk_outlier_removal(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean' : np.mean(bhk_df.price_per_feet),
                'std' : np.std(bhk_df.price_per_feet),
                'count' : bhk_df.shape[0]
            }
        
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_feet < (stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

Applying Linear Regression

```python
column_trans = make_column_transformer((OneHotEncoder(sparse=False), ['location']), remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_pred_lr,y_test)
```

Applying Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler, lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_pred_lasso,y_test)
```

Applying Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler, ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_pred_ridge,y_test)
```

```python
print("R2 SCORES ---")
print("Linear Regression :- ",r2_score(y_pred_lr,y_test))
print("Lasso Regression :- ",r2_score(y_pred_lasso,y_test))
print("Ridge Regression :- ",r2_score(y_pred_ridge,y_test))
```
